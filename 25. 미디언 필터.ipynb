```python
import sys
import numpy as np
import cv2
import matplotlib.pylab as plt
import random
!pip install opencv-python
```

```python
src = cv2.imread('junekoo.png')
src=cv2.cvtColor(src, cv2.COLOR_RGB2GRAY)
src=cv2.resize(src,(252,252))
print(src.shape)
```

```python
for i in range(0, int(src.size / 10)):
        x = random.randint(0, src.shape[1] - 1)
        y = random.randint(0, src.shape[0] - 1)
        src[x, y] = (i % 2) * 255

dst1 = cv2.GaussianBlur(src, (0, 0), 1)
dst2 = cv2.medianBlur(src, 3)
dst3 = cv2.medianBlur(dst2, 3)
#plt.figure(figsize=(12, 3))
plt.subplot(2,2,1),plt.axis('off'),plt.title('original')
plt.imshow(src,cmap='gray')
plt.subplot(2,2,2),plt.axis('off'),plt.title('GaussianBlur')
plt.imshow(dst1,cmap='gray')
plt.subplot(2,2,3),plt.axis('off'),plt.title('medianBlur')
plt.imshow(dst2,cmap='gray')
plt.subplot(2,2,4),plt.axis('off'),plt.title('medianBlur x 2')
plt.imshow(dst3,cmap='gray')
```
