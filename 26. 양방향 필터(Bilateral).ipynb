```python
import sys
import numpy as np
import cv2
import matplotlib.pylab as plt
!pip install opencv-python
```

```python
src = cv2.imread('junekoo.png')
src=cv2.cvtColor(src, cv2.COLOR_RGB2BGR)
```

```python
noise = np.zeros(src.shape, np.int32)
cv2.randn(noise, 0, 5)
cv2.add(src, noise, src, dtype=cv2.CV_8UC1)

dst1 = cv2.GaussianBlur(src, (0, 0), 5)
dst2 = cv2.bilateralFilter(src, -1, 10, 5)
    
#plt.figure(figsize=(12, 3))
plt.subplot(1,3,1)
plt.imshow(src),plt.title('original')
plt.subplot(1,3,2)
plt.imshow(dst1),plt.title('GaussianBlue')
plt.subplot(1,3,3)
plt.imshow(dst2),plt.title('bilateralFilter')
```
