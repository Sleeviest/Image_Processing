```python
src = cv2.imread('junekoo.png', cv2.IMREAD_GRAYSCALE)

emboss = np.array([[-1, -1, 0],
                   [-1, 0, 1],
                   [0, 1, 1]], np.float32)

dst = cv2.filter2D(src, -1, emboss, delta=128)

#plt.figure(figsize=(12, 3))
plt.subplot(1,2,1),plt.imshow(src,cmap='gray')
plt.subplot(1,2,2),plt.imshow(dst,cmap='gray')
```
